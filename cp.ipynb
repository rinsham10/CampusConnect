```python
x=df.drop(['status'],axis=1)
y1=df['status']
y=pd.DataFrame(y1)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='rbf')
```

```python
model=svm.fit(x_train,y_train)
pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred))
```

```python
# Features in the same order as in training
student_data = {
    'ssc_p': 61.8,        # SSC percentage
    'hsc_p': 47.0,        # HSC percentage
    'degree_p': 54.38,     # Degree percentage
    'workex': 'No',      # 'Yes' or 'No'
    'etest_p': 57.00,      # Employability test score
    'mba_p': 56.13         # MBA percentage
}

student_data['workex'] = 1 if student_data['workex'].strip().lower() == 'yes' else 0


input_df = pd.DataFrame([student_data])[['ssc_p', 'hsc_p', 'degree_p', 'workex', 'etest_p', 'mba_p']]

prediction = model.predict(input_df)[0]

# 
if prediction == 1:   # Assuming 1 = Placed, 0 = Not Placed
    print("✅ The student is Placed.")
else:
    print("❌ The student is Not Placed.")
```
